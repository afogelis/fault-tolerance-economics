# Breaking RSA-2048: A Physical-Qubit Estimate

How many physical qubits and how much time would Shor's algorithm need to factor a 2048-bit RSA
modulus under realistic surface-code error correction? We estimate across hardware profiles and
analyse the sensitivity to physical error rate, cycle time and threshold. The baseline reproduces
the order of magnitude of Gidney & Ekera (2021): ~20 million qubits in ~8 hours.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

from fteconomics import DEFAULT_PROFILES, SHOR_RSA_2048, estimate_resources
from fteconomics.viz import plot_sensitivity

rows = []
for name, profile in DEFAULT_PROFILES.items():
    est = estimate_resources(SHOR_RSA_2048, profile)
    rows.append(
        {
            "profile": name,
            "distance": est.code_distance,
            "physical_qubits_millions": round(est.total_physical_qubits / 1e6, 1),
            "runtime_hours": round(est.runtime_hours, 1),
        }
    )
pd.DataFrame(rows)
```

```python
plot_sensitivity(SHOR_RSA_2048, DEFAULT_PROFILES["baseline"])
plt.tight_layout()
plt.show()
```
